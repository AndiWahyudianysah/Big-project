```python
# Import Library yang akan digunakan
import pandas as pd

# Membaca datasets menggunakan perintah pd.read_csv
data = pd.read_csv('/content/drive/MyDrive/File Google Collab/Mushroom Dataset.csv')
data
```

```python
#menghilangkan variabel yang tidak dibutuhkan
columns_to_drop = ['does-bruise-or-bleed','gill-attachment','gill-spacing','gill-color','ring-type']
data = data.drop(columns=columns_to_drop)
data
```

```python
#Visualisasi Datasets dengan menggunakan Boxplot (data numerik)
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure()
sns.boxplot(data)
```

```python
#Visualisasi Datasets variabel kategorik dengan 2 kategori menggunakan Pie Chart

# Import Library yang akan digunakan
import pandas as pd
import matplotlib.pyplot as plt

# Variabel Kategorik dengan 2 kategori
kategori2 = ['class','has-ring']

# Looping untuk membuat pie chart untuk setiap kategori
for kategori in kategori2:
    # Hitung jumlah masing-masing kategori
    kategori_counts = data[kategori].value_counts()

    # Buat pie chart
    plt.figure()
    plt.pie(kategori_counts, labels=kategori_counts.index, autopct='%1.1f%%', startangle=90)
    plt.title(kategori, y=0.92)
    plt.show()
```

```python
#Visualisasi Datasets variabel kategorik dengan > 2 kategori menggunakan Barplot

# Import Library yang akan digunakan
import pandas as pd
import matplotlib.pyplot as plt

kategorik = ['cap-shape','cap-surface','cap-color','stem-color','habitat','season']

# Mengatur jumlah baris dan kolom untuk subplot
num_rows = len(kategorik) // 3 + (1 if len(kategorik) % 3 != 0 else 0)
num_cols = 3

# Loop untuk membuat plot untuk setiap variabel
fig, axs = plt.subplots(num_rows, num_cols, figsize=(15, 3 * num_rows))
for i, variable in enumerate(kategorik):
    row = i // num_cols
    col = i % num_cols

    # Menghitung frekuensi
    value_counts = data[variable].value_counts()

    # Membuat barplot dengan warna yang ditentukan
    axs[row, col].bar(value_counts.index, value_counts.values, color='orange')

    # Menambahkan teks di atas setiap bar
    for index, value in zip(value_counts.index, value_counts.values):
        axs[row, col].text(index, value + 1, str(value), ha='center', fontsize=7)

    axs[row, col].set_title(variable)  # Judul plot
    axs[row, col].set_xlabel(variable)  # Label sumbu x
    axs[row, col].set_ylabel('Frekuensi')  # Label sumbu y

# Menyembunyikan subplot yang tidak terpakai
for i in range(len(kategorik), num_rows * num_cols):
    row = i // num_cols
    col = i % num_cols
    fig.delaxes(axs[row, col])

plt.tight_layout()
plt.show()
```

**One Hot Encoding**


```python
# Import Library yang akan digunakan
import numpy as np

max_k = 13 # Jumlah kategori paling banyak pada variabel yang akan di encode

# Daftar kolom yang akan dilakukan one-hot encoding
columns_to_encode = columns_to_encode = ['class','cap-shape','cap-surface','cap-color','stem-color','has-ring','habitat','season']

# Lakukan one-hot encoding untuk setiap kolom
for col in columns_to_encode:
    top_values = data[col].value_counts().nlargest(max_k).index
    for value in top_values:
        data[col + '_' + str(value)] = (data[col] == value).astype(int)

# Hapus kolom asli yang telah di-encode
data_encoded = data.drop(columns=columns_to_encode)
data_encoded
```

**Feature Scaling**

```python
# Impor Library yang akan digunakan
import pandas as pd
from sklearn.preprocessing import RobustScaler

# Inisialisasi MinMaxScaler
scaler = RobustScaler()

# Memilih kolom yang ingin di-scaling
kolom_scaling = ['cap-diameter','stem-height','stem-width']

# Melakukan scaling pada kolom yang telah ditentukan
data_encoded[kolom_scaling] = scaler.fit_transform(data_encoded[kolom_scaling])
data_scaling = data_encoded
data_scaling
```

**Clustering**

```python
# Menentukan data yang akan diklusterkan
data_cluster = data_scaling[['cap-diameter','stem-height','stem-width']]
data_cluster
```

```python
# Menentukan jumlah cluster dengan Elbow
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Buat dataset menjadi array
X = np.asarray(data_cluster)

# Inisialisasi nilai K yang mungkin
k_values = range(1, 11)
# Hitung inertia (WCSS) untuk setiap nilai K
inertia = []
for k in k_values:
    kmeans = KMeans(n_clusters=k, n_init=10)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# Plot hasil elbow analysis
plt.plot(k_values, inertia, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('WCSS (Inertia)')
plt.title('Elbow Method for Optimal k')
plt.xticks(k_values)
plt.show()
```

```python
# Menentukan jumlah cluster dengan silhouette score
from sklearn.metrics import silhouette_score
score = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, n_init=10)
    label = kmeans.fit_predict(X)

    siluet = silhouette_score(X, label, metric='euclidean')
    score.append(siluet)

plt.figure(figsize=(10,5))
plt.plot(range(2, 10), score)
plt.ylabel('Silhouette')
plt.xlabel('n_clusters')
```

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Mempersiapkan data untuk clustering
X = np.asarray(data_cluster)

# Melakukan clustering dengan KMeans
kmeans = KMeans(n_clusters=2, n_init=10)
label = kmeans.fit_predict(X)
centroid = kmeans.cluster_centers_

# Memvisualisasikan hasil clustering
plt.figure(figsize=(6,6))
plt.scatter(X[:, 0], X[:, 1], c=label, s=10)
plt.scatter(centroid[:, 0], centroid[:, 1], c='white', edgecolors='k', s=100, linewidths=2)
plt.axis('equal');
```

**Klasifikasi**

```python
# Spliting Dataset
from sklearn.model_selection import train_test_split

# Mengambil fitur dan label dari dataset
X = data_scaling.drop(columns=['class_p', 'class_e'])
y = data_scaling[['class_p', 'class_e']]

# Membagi dataset menjadi data training+validation dan data test
X_train_val, X_test, y_train_val, y_test = train_test_split(X, y, test_size=0.1, random_state=0)
# Membagi data training+validation menjadi data training dan data validation
X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, test_size=0.1, random_state=0)

# Menampilkan bentuk dari data latih dan data uji
print("Bentuk data latih (X_train):", X_train.shape)
print("Bentuk data validasi (X_val):", X_val.shape)
print("Bentuk data uji (X_test):", X_test.shape)
print("Bentuk target data latih (y_train):", y_train.shape)
print("Bentuk target data validasi (y_val):", y_val.shape)
print("Bentuk target data uji (y_test):", y_test.shape)
```

**Klasifikasi dengan ANN**

```python
import tensorflow as tf

# Membangun model ANN
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(X.shape[1],)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(2, activation='sigmoid')
])

# Mengkompilasi model dengan binary crossentropy dan optimizer Adam
model.compile(
    loss=tf.keras.losses.BinaryCrossentropy(from_logits=False),
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    metrics=['accuracy']
)

# Menampilkan ringkasan arsitektur model
model.summary()
```

```python
import tensorflow as tf

# Callback untuk early stopping
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Melatih model pada data latih
history = model.fit(X_train, y_train, epochs=100, batch_size=64, validation_data=(X_val, y_val), callbacks=[early_stopping])

# Evaluasi model pada data test
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=2)
print(f"Test accuracy: {test_accuracy:.4f}")
```

```python
import matplotlib.pyplot as plt

# Plot training & validation accuracy values
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')

# Plot training & validation loss values
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix

# Prediksi pada data test
y_pred_prob = model.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype(int)

# Classification report
print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=['class_p', 'class_e']))

# Confusion matrix
cm = confusion_matrix(y_test.values.argmax(axis=1), y_pred.argmax(axis=1))
print("Confusion Matrix:")
print(cm)

# Plot confusion matrix
import seaborn as sns
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['class_p', 'class_e'], yticklabels=['class_p', 'class_e'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

**Klasifikasi dengan Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Inisialisasi model Random Forest
clf = RandomForestClassifier(n_estimators=100, random_state=42)

# Latih model pada data training
clf.fit(X_train, y_train)
```

```python
# Evaluasi pada data validation
y_val_pred = clf.predict(X_val)
val_accuracy = accuracy_score(y_val, y_val_pred)
print(f'Validation Accuracy: {val_accuracy}')

# Evaluasi pada data test
y_test_pred = clf.predict(X_test)
test_accuracy = accuracy_score(y_test, y_test_pred)
print(f'Test Accuracy: {test_accuracy}')
```

```python
# Mengubah one-hot encoded ke label tunggal
y_test_labels = np.argmax(y_test, axis=1)
y_pred_labels = np.argmax(y_test_pred, axis=1)

# Hitung confusion matrix
cm = confusion_matrix(y_test_labels, y_pred_labels)
print("Confusion Matrix:")
print(cm)

# Hitung metrik lainnya
accuracy = accuracy_score(y_test_labels, y_pred_labels)
print("Accuracy:", accuracy)

classification_rep = classification_report(y_test_labels, y_pred_labels)
print("Classification Report:")
print(classification_rep)

# Plot confusion matrix
import seaborn as sns
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['class_p', 'class_e'], yticklabels=['class_p', 'class_e'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
